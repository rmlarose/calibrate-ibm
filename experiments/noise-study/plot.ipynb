# Plot

```python
import matplotlib.pyplot as plt; plt.rcParams.update({"font.family": "serif", "font.size": 12})
import numpy as np
```

```python
q12 = np.loadtxt("h2o_noiserates_errors_stds.txt")
q24 = np.loadtxt("2h2o_noiserates_errors_stds.txt")
q36 = np.loadtxt("3h2o_noiserates_errors_stds.txt")
```

```python
noise_rates12 = q12[:, 0]
errors12 = q12[:, 1]

noise_rates24 = q24[:, 0]
errors24 = q24[:, 1]

noise_rates36 = q36[:, 0]
errors36 = q36[:, 1]
```

```python
q14 = np.loadtxt("sto-3g_noiserates_errors.txt")
q26 = np.loadtxt("6-31g_noiserates_errors.txt")
q38 = np.loadtxt("6-311g_noiserates_errors.txt")
```

```python
# noise_rates14 = q14[:, 0]
# errors14 = q14[:, 1]

# noise_rates = q26[:, 0]
# errors = q26[:, 1]

# noise_rates2 = q38[:, 0]
# errors2 = q38[:, 1]

hardware_exact = -73.55
hardware_results = [-73.02, -72.72, -72.97]
hardware_error = np.abs(np.average(hardware_results) - hardware_exact)
hardware_error
```

```python
plt.loglog(noise_rates12, errors12, "--s", ms=9, mec="black", label=f"$n = 12$ (H$_2$O)")
plt.loglog(noise_rates24, errors24, "--s", ms=9, mec="black", label=f"$n = 24$ (2H$_2$O)")
plt.loglog(noise_rates36, errors36, "--s", ms=9, mec="black", label=f"$n = 36$ (3H$_2$O)")
plt.scatter(7e-4, hardware_error, marker="*", s=150, ec="black", label="H$_2$O on IBM Fez")
# plt.loglog(noise_rates, errors, "--s", ms=9, mec="black", label=f"$n = 26$ (6-31g)")
# plt.loglog(noise_rates2, errors2, "--s", ms=9, mec="black", label=f"$n = 38$ (6-311g)")

plt.ylabel("Absolute energy error (Ha)")
plt.xlabel("Depolarizing noise rate")
plt.legend(ncol=1)

plt.tight_layout()
plt.axhspan(0, 1e-3, color="green", alpha=0.15, zorder=0);
plt.savefig("water_vqe_scaling_nh2o.pdf")
```
