# Plot

```python
import matplotlib.pyplot as plt; plt.rcParams.update({"font.family": "serif", "font.size": 12})
import numpy as np
```

```python
q14 = np.loadtxt("sto-3g_noiserates_errors.txt")
q26 = np.loadtxt("6-31g_noiserates_errors.txt")
q38 = np.loadtxt("6-311g_noiserates_errors.txt")
```

```python
noise_rates14 = q14[:, 0]
errors14 = q14[:, 1]

noise_rates = q26[:, 0]
errors = q26[:, 1]

noise_rates2 = q38[:, 0]
errors2 = q38[:, 1]
```

```python
plt.loglog(noise_rates14, errors14, "--s", ms=9, mec="black", label=f"$n = 14$ (sto-3g)")
plt.scatter(4e-5, 0.03, marker="*", s=150, ec="black", label="$n = 14$ (sto-3g) on IBM Fez")
plt.loglog(noise_rates, errors, "--s", ms=9, mec="black", label=f"$n = 26$ (6-31g)")
plt.loglog(noise_rates2, errors2, "--s", ms=9, mec="black", label=f"$n = 38$ (6-311g)")

plt.ylabel("Absolute energy error (Ha)")
plt.xlabel("Depolarizing noise rate")
plt.legend(ncol=1)

plt.axhspan(0, 1e-3, color="green", alpha=0.15, zorder=0);
plt.tight_layout()
plt.savefig("water_vqe_scaling.pdf")
```
